#Yahoo Finance - Example

This notebook is an example about downloading and formatting data from [Yahoo Finance](http://finance.yahoo.com/) using Python and [pandas](http://pandas.pydata.org/)

```python
import numpy as np
import my_finance_tools as myfin
from datetime import datetime
from pandas import Index, DataFrame
import pandas.core.datetools as datetools
%pylab inline
```

```python
endDate = datetools.datetime.today()
startDate = endDate - datetools.DateOffset(years=2)
aapl = myfin.getQuotes('AAPL', startDate, endDate)
goog = myfin.getQuotes('GOOG', startDate, endDate)
ibm = myfin.getQuotes('IBM', startDate, endDate)
closing = DataFrame({'IBM': ibm['close'], 'GOOG': goog['close'], 'AAPL': aapl['close']})
```

```python
plt.figure()
closing.plot(label='Series', linewidth=2)
plt.legend(loc='best')
plt.grid(which='both', axis='both')
plt.title('Closing Values')
```
