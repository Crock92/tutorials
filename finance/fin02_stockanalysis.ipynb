##<a name="analysis"></a><font color="Navy">Data Analysis</font>

We are going to calculate with `pandas`:

* Log Return (**Ret**)
* Rolling mean (**rMe**)
* Rolling standard deviation (**rSD**)
* Rolling correlation between rMe and rSD (**Cor**)

```python
DAX['Ret'] = np.log(DAX['Close'] / DAX['Close'].shift(1))
DAX['rMe'] = pd.rolling_mean(DAX['Ret'], 252) * 252   
DAX['rSD'] = pd.rolling_std(DAX['Ret'], 252) * np.sqrt(252)
DAX['Cor'] = pd.rolling_corr(DAX['rMe'], DAX['rSD'], 252)  
DAX[['Open', 'Close', 'Ret', 'rMe', 'rSD', 'Cor']].ix[-10:]
```

##<a name="example1"></a><font color="Navy">A practical example: MonteCarlo Simulation</font>

**TODO: Verificare che il problema di versione di matplotlib, barh, sia risolto**

```python
AAPL['Ret'] = np.log(AAPL.Close/AAPL.Close.shift(1)) 
vol = np.std(AAPL['Ret'])*np.sqrt(252)               
r = 0.025                                            # Constant Short Rate
S0 = AAPL['Close'][-1]                               # End Value                                                                                      
K = S0 * 1.1                                         # OTM Call Option                                                                                
T = 1.0                                              # Maturity Year
M = 100; dt = T / M                                  # Time Steps                                                                                     
I = 10000                                            # Simulation Paths
S = np.zeros((M + 1, I))
S[0,:] = S0
for t in range(1, M + 1):
    ran = np.random.standard_normal(I)
    S[t, :] = S[t - 1, :] * np.exp((r - vol**2 / 2) * dt + vol * np.sqrt(dt) * ran)
montecarloPlot(S)
```

We will gather some data from an hdf5 file. By using `pandas.HDFStore` utility. These are fake data of a portfolio constructed from 3 randomly-generated factors (usually called the factor loadings) and some weights. We created the portfolio by using the formula **`0.7*fac1 - 1.2*fac2 + 0.3*fac3`** and adding some noise:

```python
h5file = pd.HDFStore(fg.get_example('p07_portfolio.h5'))
portfolio = h5file['portfolio']
factors = h5file['factors']
h5file.close()
```

```python
side_by_side(factors.head(), portfolio.head())
```

The standard way to compute the factor exposures is by **`least squares regression`**; using pandas.ols with factors as the explanatory variables we can compute exposures over the entire set of tickers:

```python
exposures = pd.ols(y=portfolio, x=factors).beta
display(exposures)
```

Now compute exposure industry by industry.      

```python
def beta_exposure(chunk, factors=None):
    return pd.ols(y=chunk, x=factors).beta
    
by_industry = portfolio.groupby(industries)
exposures = by_industry.apply(beta_exposure, factors=factors)
exposures.unstack()
```
