There are five preset seaborn themes: darkgrid, whitegrid, dark, white, and ticks.

Both the **white** and **ticks** styles can benefit from removing the top and right axes spines, which are not needed. You can call the seaborn function **despine()** to remove them:

Some plots benefit from a combined use of despine() and the function offset_spines(), which pulls the spines away from the data. When the ticks don’t cover the whole range of the axis, the trim parameter in despine() will limit the range of the surviving spines. Because of some quirks in the ways figures get drawn, you’ll get the best results if you call offset_spines() before plotting:

You can use the axes_style() function in a with statement to *temporarily* set plot parameters:

```python
with sns.axes_style("ticks"):
    plt.hist(s, 21, histtype="stepfilled", color=sns.desaturate("green", .75), alpha=0.4);
    plt.hist(s, 31, histtype="stepfilled", color=sns.desaturate("indianred", .75), alpha=0.6);
    sns.despine()
    plt.show()
```

```python
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.set_style("white")
f, ax = plt.subplots()
sns.offset_spines()
sns.violinplot(data)
sns.despine(trim=True);
```

To grab the current color cycle, call the color_palette() function with no arguments. This just returns a list of r, g, b tuples:

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

There are six variations of the default theme, called pastel, bold, muted, deep, dark, and colorblind:

```python
sns.palplot(sns.color_palette("pastel"))
```

If you request more colors than exist for a given qualitative palette, the colors will cycle:

```python
sns.palplot(sns.color_palette("deep", 15))
```

*For Palettes you can also use the name of any matplotlib colormap names*, and the palette will return evenly-spaced samples from points near the extremes.

```python
sns.palplot(sns.color_palette("cool", 15))
```

You can just pass in a list of color codes to specify a custom palette:

```python
sns.palplot(sns.color_palette(["#9b59b6", "#3498db", "#e74c3c", "#34495e", "#2ecc71"]))
```

**dark_palette** creates a blend to it from a very dark gray. The interpolation is not currently performed in a color space that is compatible with human perception, so the increments of color in these palettes will not necessarily appear uniform.

*For Colornames use [CSS colornames](http://www.w3schools.com/cssref/css_colornames.asp):*

```python
sns.palplot(sns.dark_palette("Aqua", 12, reverse=True))
```

There’s a related trick embedded in palette production that allows you to specify the name of a sequential ColorBrewer palette with a “_d” suffix. This will create a palette that is harmonious with the base palette, but using colors that are dark enough to draw line or contour plots.

A more general function for making custom palettes is **blend_palette()**, which interpolates between an arbitrary number of seed points. You could use this to make your own diverging palette:

```python
sns.palplot(sns.blend_palette(["CornflowerBlue", "AliceBlue", "Coral"], 11))
```

**desaturate()** defines a % in color desaturation

```python
sns.palplot(sns.blend_palette(["CornflowerBlue", "AliceBlue",
                               sns.desaturate("#FF7F50", 0.5)], 11))
```

Any seaborn function that draws plots with multiple color is using color_palette() behind the scenes, and will accept a palette for the color or palette argument in any of the ways shown above:

```python
pal = sns.blend_palette(["seagreen", "lightblue"], 3)
sns.palplot(pal)
```

```python
tips = sns.load_dataset("tips")
sns.factorplot("day", "total_bill", hue="smoker", data=tips, kind="box");
```

```python
sns.factorplot("day", "total_bill", hue="smoker", data=tips, kind="box", palette=pal);
```

```python
with sns.color_palette(pal):
    sns.factorplot("day", "total_bill", hue="smoker", data=tips, kind="box");
```

```python
from scipy import stats
```

```python
data1 = stats.poisson(2).rvs(300)
data2 = stats.poisson(5).rvs(2000)
max_data = np.r_[data1, data2].max()
bins = np.linspace(0, max_data, max_data + 1)
plt.hist(data1, bins, normed=True, color="#6495ED", alpha=.5)
plt.hist(data2, bins, normed=True, color="#F08080", alpha=.5);
```

```python
x = stats.gamma(3).rvs(3000)
y = stats.gamma(5).rvs(3000)
sns.jointplot(x, y, kind='hex');
```
