```python
def _string_param(params):
    p_string = "\nParams: \n"
    for k in params.keys():
        p_string = p_string + str(k) + str(params[k]) + "\n"
    return p_string

def _random_int_list(start,end,length):
    i = 0
    l = []
    while i<length:
        x = randint(start,end) #Inclusive
        if x not in l:
            i = i + 1
            l.append(x)
    return l
```

```python
features_list_readable = []
features_list = []
min_features = 5
max_features = 20

last_data_col = len(list(df.ix[:,:].columns))-1
x = find_last_data(df)
if x>-1:
    last_data_col = x
    
full_list = ['NONE']+['ALL']+list(df.ix[:,:last_data_col+1].columns)
actions = ['NONE']+['RESET']+['ADD']
multiple_features_lists = []
multiple_features_lists_2 = []
multiple_features_lists_3 = []
pseudo_random = False
low_priority_list = False

def equals(l1, l2):
    if len(l1) == len(l2):
        for i in l1:
            if i not in l2:
                return False
    else:
        return False
    return True

def list_exists(l):
    for i in multiple_features_lists:
        if equals(l,i):
            return True
    for i in multiple_features_lists_2:
        if equals(l,i):
            return True
    for i in multiple_features_lists_3:
        if equals(l,i):
            return True
    return False

def features_translation(fl):
    translation = []
    for i in fl:
        translation.append(full_list[i+2])
    return translation

def features_selection(c='NONE'):
    if c != 'NONE' and c!= 'ALL' and c not in features_list_readable:
        features_list_readable.append(c.strip())
        features_list.append(full_list.index(c)-2)
        print " , ".join(features_list_readable)
    elif c == 'ALL':
        for i in full_list:
            if i != 'NONE' and i != 'ALL' and i != 'label' and \
               i != 'Label' and i not in features_list_readable:
                features_list_readable.append(i.strip())
                features_list.append(full_list.index(i)-2)
        print " , ".join(features_list_readable)
    return 

def reset():
    features_list_readable[:] = []
    features_list[:] = []
    return

def print_lists():
    print "High priority lists"
    for i in multiple_features_lists:
        print features_translation(i)
    print "Medium priority lists"
    for i in multiple_features_lists_2:
        print features_translation(i)
    print "Low priority lists"
    for i in multiple_features_lists_3:
        print features_translation(i)
    return

def expand_list(l):
    if len(l)== max_features or len(l)==len(full_list)-2:
        return []
    expanded_list = []
    
    for i in full_list:
        if i not in ['NONE','ALL'] and i not in features_translation(l):
            new_l = list(l)
            new_l.append(full_list.index(i)-2)
            new_l.sort()
            if list_exists(new_l)==False:
                expanded_list.append(new_l)
                
    temp = list(expanded_list)
    for i in temp:
        temp2 = expand_list(i)
        if len(temp2)>0:
            expanded_list.extend(temp2)
    return expanded_list

def pseudo_random_lists(l):
    expanded_lists = expand_list(l)
    for i in expanded_lists:
        if len(i)>0 and len(i)>=min_features and len(i)<=max_features and list_exists(i)==False:
            if low_priority_list==False:
                multiple_features_lists_2.append(i)
            else:
                multiple_features_lists_3.append(i)
    if low_priority_list==True:
        print_lists()
    return

def selection_completed(d='NONE'):
    if d=='NONE':
        return
    elif d == 'ADD' and len(features_list)>0:
        ordered_list = features_list[:]
        ordered_list.sort()
        if list_exists(ordered_list)==False:
            if pseudo_random==False:
                multiple_features_lists.append(ordered_list)
            else:
                pseudo_random_lists(ordered_list)
            print 'Current list added'
        else:
            print 'List already added!'
        reset()
    elif d == 'RESET' and len(features_list)>0:
        reset()
        print 'List resetted'
    print "Lists saved:\n"
    print_lists()
    return 
```

```python
masks_list = []
for i in multiple_features_lists[0]:
         masks_list.append(i+1+last_data_col)
#df_m = df.ix[:,masks_list]
df_m = (df[df[masks_list]>0.5]).sum(axis=1)
#df[df['BoolCol'] == True].index.tolist()
#print (df[df[masks_list]>0.5]).sum(axis=1)
#print df_m
df_m = df_m[df_m==len(multiple_features_lists[0])]
print df_m.index.tolist()
```

```python
def get_df_mask(features_l):
    masks_list = []
    for i in features_l:
         masks_list.append(i+1+last_data_col)
    df_m = (df[df[masks_list]>0.5]).sum(axis=1)
    df_m = df_m[df_m==len(features_l)]
    #print df_m.index.tolist()
    return df_m.index.tolist()

def apply_masks(features_l):
    row_to_keep = get_df_mask(features_l)
    target = df.ix[row_to_keep,-1]
    data = df.ix[row_to_keep,features_l]
    return target, data
```

```python
def data_extraction(features_l):
    target, data = apply_masks(features_l)
    idx_train, idx_valid = cross_validation.train_test_split(data.index, test_size=0.15)
    df_train, df_valid = data.ix[idx_train], data.ix[idx_valid]
    y_train, y_valid = target[idx_train],target[idx_valid]
    return df_train, df_valid, y_train, y_valid, data, target
```

```python
#algorithm=0 -> gtb
#algorithm=1 -> rfc
def test(features, params, algorithm, test_count):
    log_file.write("="*50)
    log_file.write("\nTest counter: " + str(test_count))
    
    if algorithm==1:
        log_file.write("\nAlgorithm : RandomForestClassifier\n")
    else:
        log_file.write("\nAlgorithm : GradientBoostingClassifier\n")
    print "Data extraction..."
    p_string = _string_param(params)
    df_train, df_valid, y_train, y_valid, data, target = data_extraction(features)
    
    
    
    log_file.writelines(p_string)
    log_file.flush()
    
    alg = ensemble.GradientBoostingClassifier()
    if(algorithm==1):
        alg = ensemble.RandomForestClassifier()
    
    #grid = grid_search.GridSearchCV(rfc, params, cv=3, n_jobs=1)
    grid = grid_search.GridSearchCV(alg, params, n_jobs=-1)
    log_file.write("\nGrid Search result:\n" + str(grid))
    log_file.flush()
    t0 = time()
    print "Training phase..."
    grid.fit(df_train, y_train)
    fit_time = time() - t0
    best = grid.best_estimator_
    print "Validation phase"
    out1 = metrics.confusion_matrix(best.predict(df_valid), y_valid)
    if algorithm==1:
        out2 = '\nBest Params: N est:%3i - Mx feat:%2i - Mx dpth:%2i - Mn s.splt:%2i - Mn s.leaf:%2i - Bootstrap:%2i - F1:%.3f'
        %(best.n_estimators, best.max_features, best.max_depth, best.min_samples_split, best.min_samples_leaf, best.bootstrap, \
           metrics.f1_score(best.predict(df_valid), y_valid))
    else:
        out2 = '\nBest Params: N est:%3i - Mx feat:%2i - Mx dpth:%2i - Mn s.splt:%2i - Mn s.leaf:%2i - F1:%.3f' 
        %(best.n_estimators, best.max_features, best.max_depth, best.min_samples_split, best.min_samples_leaf, \
           metrics.f1_score(best.predict(df_valid), y_valid))
    out3 = '\nFit time %0.3f[s]' %fit_time

    #UPDATE LOGFILE
    log_file.write("\n\nConfusion matrix: \n" + str(out1) + "\n")
    log_file.write(out2)
    log_file.write(out3 + "\n\n")
    log_file.flush()
    best.feature_importances_
    feature_importance = best.feature_importances_
    feature_importance = 100.0*(feature_importance/feature_importance.max())
    feature_importance_sorted_idx = np.argsort(feature_importance)
  
    feature_importance = best.feature_importances_
    feature_importance_sorted_idx = np.argsort(feature_importance)
    if algorithm==1:
        log_file.write("\n\n\FEATURES IMPORTANCE RFC\n")
    else:
        log_file.write("\n\n\FEATURES IMPORTANCE BTG\n")
    for i in reversed(feature_importance_sorted_idx):
        log_file.write(str(data.columns[i]) + "  " + str(feature_importance[i])+"\n")
    log_file.flush()

    return
```
