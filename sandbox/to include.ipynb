```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 3*np.pi, 700)
plt.figure(figsize=(17, 4))
plt.plot(x, np.sin(x**2), linewidth=2, color='magenta')
plt.title(r'A simple Chirp: $sin(x^{2})$', fontsize=18);
```

```python
def edit_myplot(lw=1.0, n=2.0):
    plt.figure(figsize=(17, 4))
    plt.plot(x, np.sin(x**n), linewidth=lw, color='magenta')
    plt.title('An interactive Chirp', fontsize=18);
    return
```

```python
from IPython.html.widgets import interact, fixed
interact(edit_myplot, lw=(0.5, 5, 0.5), n=(0.5, 3, 0.1));
```
