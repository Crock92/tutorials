```python
start = datetime.datetime(2000, 1, 1)
end = datetime.datetime(2014, 5, 29)
AAPL = web.DataReader("AAPL", 'yahoo', start, end)
#IBM_yahoo = web.DataReader("IBM", 'yahoo', start, end)
```

Plot Open and Closing prices of the last 'n' transactions:

```python
import matplotlib.gridspec as gridspec
fig = plt.figure(figsize=(15,10));
gs = gridspec.GridSpec(2, 2, height_ratios=[2,1], hspace=0.4)
ax0 = plt.subplot(gs[0, :])
ax1 = plt.subplot(gs[1, :])
AAPL.iloc[-200:,1:3].plot(ax=ax0);
AAPL['r_mean'] = pd.rolling_mean(AAPL.ix[:,'High'],20)
AAPL.ix[-200:,['High', 'r_mean']].plot(ax=ax1);
```

```python
df = pd.DataFrame(np.random.randn(1000, 6), columns=list('ABCDEF'),
                  index = pd.date_range('1/1/2000', periods=1000))
df.hist(color=(0.2,0.0,0.8), alpha=0.6, bins=25, sharex=True,
        layout=(2,3), figsize=(15,6), grid=False);
```

DataFrame has a boxplot() method that allows you to visualize the distribution of values within each column:

```python
df.boxplot(figsize=(15,6));
```

```python
df_grouped = df.copy()
df_grouped['group'] = 0.0
df_grouped['group'].ix[(df_grouped['B']>0) & (df_grouped['C']<0)] = 1.0
df_grouped.boxplot(figsize=(15,12), by=['group']);
```
